# Example of Neural Network:

```python
# Inputs
x = torch.tensor(6.7) # input feature
y = torch.tensor(0.0) # true label

w = torch.tensor(1.0) # weight
b = torch.tensor(0.0) # bias
```

```python
# Binary Cross-Entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
  epsilon = 1e-8
  prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
  return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
```

```python
# Forward Pass
z = w * x + b # weighted sum
y_pred = torch.sigmoid(z) # predicted probability

# Compute binary Cross_entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
```

```python
# Backward Pass
# Derivatives
# 1. dL/d(y_pred)
dloss_dy_pred = (y_pred-y)/(y_pred*(1-y_pred))

# 2. dy_pred/dz
dy_pred_dz = y_pred*(1-y_pred)

# 3. dz/dw & dz/db
dz_dw = x
dz_db = 1

# Compute Gradients
dloss_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dloss_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"dloss_dw: {dloss_dw}") # w.r.t weight
print(f"dloss_db: {dloss_db}") # w.r.t bias
```

```python
x = torch.tensor(6.7) # input feature
y = torch.tensor(0.0) # true label

w = torch.tensor(1.0, requires_grad=True) # weight
b = torch.tensor(0.0, requires_grad=True) # bias
```

```python
print(w)
print(b)
```

```python
z = w * x + b # weighted sum
y_pred = torch.sigmoid(z) # predicted probability

# Compute binary Cross_entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
loss
```

```python
loss.backward()
```
