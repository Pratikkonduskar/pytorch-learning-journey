# Reduction Operations:

```python
e = torch.randint(size=(2,3), low=0, high=10)
print(e)
```

```python
# sum
print(torch.sum(e))
# sum across columns
print(torch.sum(e, dim=0))
# sum across rows
print(torch.sum(e, dim=1))
```

```python
# mean
print(torch.mean(e, dtype=float))
# mean across columns
print(torch.mean(e, dim=0, dtype=float))
# mean across rows
print(torch.mean(e, dim=1, dtype= float))
```

```python
# median
print(torch.median(e))
# median across columns
print(torch.median(e, dim=0))
# median across rows
print(torch.median(e, dim=1))
```

```python
# max and min
print(torch.max(e))
print(torch.min(e))
```

```python
# prod
print(torch.prod(e))
# prod across columns
print(torch.prod(e, dim=0))
# prod across rows
print(torch.prod(e, dim=1))
```

```python
torch.std(e.float())
```

```python
# variance
torch.var(e.float())
```

```python
# argmax
torch.argmax(e)
```

```python
# agrmin
torch.argmin(e)
```

# Tensor operations on GPU:

```python
torch.cuda.is_available()
```

```python
device= torch.device('cuda')
```

```python
# creating a new tensor on GPU
a = torch.rand((2,3), device=device)
a
```

```python
# moving an existing tensor to GPU
a=torch.rand(2,3)
a
```

```python
b=a.to(device)
```

```python
b
```

```python
b+5
```

```python
import time
size=10000
#create random matrix on CPU
a=torch.randn(size,size)
b=torch.randn(size,size)

# Measure time on CPU
start=time.time()
c=torch.matmul(a,b)
end=time.time()
y=end-start
print(y)

# move matrix to GPU
c=a.to(device=device)
d=b.to(device=device)

# measure time on GPU
start=time.time()
e=torch.matmul(c,d)
end=time.time()
x=end-start
print(x)
# compare results
print("CPU time/GPU time: ", y/x)
```
